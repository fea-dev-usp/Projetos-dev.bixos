# Importando depósitos à vista

```python
import pandas as pd

depositos_link = 'http://api.bcb.gov.br/dados/serie/bcdata.sgs.27790/dados?formato=json'
depositos_a_vista = pd.read_json(depositos_link)

valor_depositos = depositos_a_vista['valor']
data_depositos = depositos_a_vista['data']

print(depositos_a_vista)
```

# Importando papel moeda em poder do público do BC

```python
pmpp_link = 'http://api.bcb.gov.br/dados/serie/bcdata.sgs.27789/dados?formato=json'
pmpp = pd.read_json(pmpp_link)

valor_pmpp = pmpp['valor']
data_pmpp = pmpp['data']

print(pmpp)
```

# Relacionando os dados de pmpp x deposito à vista (Meios de Pagamento) com o lançamento da nota de 200,00 reais

```python
import pandas as pd
import numpy as np
import datetime

#depositos
depositos = pd.DataFrame({'Joined_date': pd.to_datetime(data_depositos, format = '%d/%m/%Y'),"valor":valor_depositos})

filtered_depositos = depositos.query("Joined_date >= '2020-01-1' and Joined_date <='2021-12-31'")
print(filtered_depositos)

#pmpp
pmpp = pd.DataFrame({'Joined_date': pd.to_datetime(data_pmpp, format = '%d/%m/%Y'),"valor":valor_pmpp})

filtered_pmpp = pmpp.query("Joined_date >= '2020-01-1' and Joined_date <='2021-12-31'")
print(filtered_pmpp)
```

```python
import matplotlib.pyplot as plt


#tamanho do gráfico
plt.rcParams['figure.figsize'] = [11, 7]
plt.style.use('fivethirtyeight')
fig = plt.figure()


#dando nome as variáveis x e y
depositosd_filtered = filtered_depositos['Joined_date']
depositosv_filtered = filtered_depositos['valor']

pmppd_filtered = filtered_pmpp['Joined_date']
pmppv_filtered = filtered_pmpp['valor']

plt.plot(depositosd_filtered, depositosv_filtered, color = 'g', label = 'Depósitos à vista' )
plt.plot(pmppd_filtered, pmppv_filtered, color = 'r', label = 'Papel moeda em poder do público')

#definindo título,
plt.title('Evolução dos meios de pagamento com o lançamento da nota de 200 reais')

plt.legend()
plt.show()
```

# Customizando o gráfico

# Meios de Pagamento 

```python
import pandas as pd

Meios_de_pagamento_link = 'http://api.bcb.gov.br/dados/serie/bcdata.sgs.27841/dados?formato=json'
Meios_de_pagamento = pd.read_json(Meios_de_pagamento_link)

Valor_Meios_de_pagamento = Meios_de_pagamento['valor']
Data_Meios_de_pagamento = Meios_de_pagamento['data']

print(Meios_de_pagamento)
```

```python
import pandas as pd
import numpy as np
import datetime


MP = pd.DataFrame({'Joined_date': pd.to_datetime(Data_Meios_de_pagamento, format = '%d/%m/%Y'),"valor":Valor_Meios_de_pagamento})

filtered_MP = MP.query("Joined_date >= '2020-01-1' and Joined_date <='2021-12-31'")
print(filtered_MP)
```

# Mudando a cor de acordo com o crescimento

```python
dataMP_filtered = filtered_MP['Joined_date']
valorMP_filtered = filtered_MP['valor']


def cor_do_ponto(valor: float) -> str:
    """Retorna a cor de um ponto do gráfico, de acordo com o seu valor."""
    if valor < 437758208:
        return "royalblue"
    elif valor < 536925421:
        return 'limegreen'
    elif valor <  611838801:
         return 'orange'
    else:
           return 'red'
        

# Cria uma nova coluna aplicando a função a cada ponto da coluna Status
filtered_MP['Color'] = valorMP_filtered.apply(cor_do_ponto)


# Plota o resultado
plt.scatter(dataMP_filtered, valorMP_filtered, color = filtered_MP['Color'])


plt.show()
```
